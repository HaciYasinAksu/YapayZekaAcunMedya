Elbows metotunda her k means modelinde inertia deeri vardr. Dirsek noktasn bulmak ok nemli  

```python
def elbow_method():
    inertias=[]
    for k in range(2,11): 
        model = KMeans(n_clusters=k,random_state=42)
        model.fit(X)
        inertias.append(model.inertia_)
    plt.figure(figsize=(5,3))
    plt.plot(range(2,11),inertias)
    print(inertias)
```

```python
elbow_method()
```

elbow metot her zaman kendi noktama olan uzakla bakar lakin dier noktalarn merkez noktalarna da bakmamz lazm . Bunu da siluet metot hem kendi merkeze yaknlk hem de dier merkezlere olan uzaklklara dikkat eder.


```python
from sklearn.metrics import silhouette_score
```

```python
def silh_method():
    scores =[] 
    for k in range (2,20):
        model =KMeans(n_clusters=k,random_state=42)
        y = model.fit_predict(X)
        scores.append( silhouette_score(X,y))
    plt.figure(figsize=(5,11))
    plt.plot(range(2,20),scores)
    plt.show()
```

```python
silh_method()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score

model = KMeans(n_clusters=4, random_state=42)
model.fit(X)

# Tahmin edilen etiketleri al
labels = model.labels_

# Doru ekilde DBI hesapla
dbi = davies_bouldin_score(X, labels)

"Davies-Bouldin Index:", dbi
```
